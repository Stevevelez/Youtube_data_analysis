## **Viral** **channnel**

```python
grouped_by_channel_df = main_df.groupby(['channel_title']).sum()
grouped_by_channel_df.head(20)

viral_df = grouped_by_channel_df.sort_values(
by="views",
ascending=False
)

viral_df.head(50)
```

## **Polemic** **channel**

```python
polemical_grouped_by_channel_df = grouped_by_channel_df.copy()

polemical_grouped_by_channel_df = polemical_grouped_by_channel_df[polemical_grouped_by_channel_df['likes'] != 0]
polemical_grouped_by_channel_df = polemical_grouped_by_channel_df[polemical_grouped_by_channel_df['dislikes'] != 0]

polemical = []

all_likes = polemical_grouped_by_channel_df['likes'].tolist()
all_dislikes = polemical_grouped_by_channel_df['dislikes'].tolist()

for i in range(len(all_likes)):

  polemical.append(all_dislikes[i]/all_likes[i])

pol = pd.Series(polemical)



polemical_grouped_by_channel_df['polemic'] = pol.values

polemical_grouped_by_channel_df = polemical_grouped_by_channel_df[(polemical_grouped_by_channel_df['polemic'] >= 0.8) & (polemical_grouped_by_channel_df['polemic'] <= 1.2)]


polemical_grouped_by_channel_df = polemical_grouped_by_channel_df.sort_values(
by="comment_count",
ascending=False
)

polemical_grouped_by_channel_df.head(10)
```

## **Specific session content**

```python
specific_section_df = main_df.groupby(['channel_title', 'tags'], as_index=False).sum()

specific_section_df = specific_section_df.sort_values(
by="views",
ascending=False
)

specific_section_df = specific_section_df.iloc[0:9999]
desired_section = "books"

recommended_channels_df = specific_section_df[specific_section_df['tags'].str.contains(desired_section)]
recommended_channels_for_section = list(recommended_channels_df["channel_title"])
recommended_channels_for_section = list(dict.fromkeys(recommended_channels_for_section))

print(recommended_channels_for_section)
```

## **One** **hit** **wonder**

```python
one_hit_wonder_df = main_df.groupby(['title', 'channel_title'], as_index=False).sum()

one_hit_wonder_df = one_hit_wonder_df.sort_values(
by="views",
ascending=False
)

one_hit_wonder_df.head(10)
```

```python
channel_titles = list(one_hit_wonder_df["channel_title"].iloc[0:9])

for ct in channel_titles:
  hits = one_hit_wonder_df[(one_hit_wonder_df['channel_title'] == ct) & (one_hit_wonder_df['views'] > 1000000)]
  number_of_channel_hits = len(hits)

  print(f"Number of channel hits {number_of_channel_hits}")

  if number_of_channel_hits == 1:
    print(f"Artist/Channel {ct} is a One Hit Wonder \n")
```

## **Best tags**

```python
best_tags_df = main_df.groupby(['channel_title', 'tags'], as_index=False).sum()

best_tags_df = best_tags_df.sort_values(
by="views",
ascending=False
)

best_tags_df = best_tags_df.iloc[0:5]
best_tags_strings = list(best_tags_df["tags"])
best_tags = []

for tags_string in best_tags_strings:
  tag_list = []

  for tag in tags_string.split("|"):
    tag_list.append(tag.replace('"', ""))

  best_tags += tag_list

best_tags = list(dict.fromkeys(best_tags))
best_tags
```

## **Tittle features must-haves**

```python
best_title_df = main_df.groupby(['title'], as_index=False).sum()

best_title_df = best_title_df.sort_values(
by="views",
ascending=False
)

best_title_df = best_title_df.iloc[0:9]
best_titles = list(best_title_df["title"])

number_of_chars = []
number_of_words = []
best_words_to_use = []

for title in best_titles:
  number_of_chars.append(len(title))
  words_list = title.replace(",","").replace(".","").split(" ")

  if "|" in words_list:
     words_list.remove("|")
  if "-" in words_list:
    words_list.remove("-")

  number_of_words.append(len(words_list))
  best_words_to_use += words_list

best_words_to_use = list(dict.fromkeys(best_words_to_use))

print(f"AVG Number of Chars {round(mean(number_of_chars))}")
print(f"AVG Number of Words {round(mean(number_of_words))}")
print(f"Best Words to use in Title {best_words_to_use}")
```
